Facebook AI Similarity Search(Faiss) is a library for efficient similirity search and clusterinng of dense vectors. It contains algorithms that search in sets of vectors of any size , up to ones that possibly do not fit in RAM. It also contains supporting code for evaluation and parameter tunings

```python
# Install the required package

from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter 
```

```python
# Load the text file using TextLoader
loader = TextLoader(r"D:\LangChain Project\VectorStore\speech.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="gemma:2b")
db=FAISS.from_documents(docs,embeddings)
db
```

```python
## qurering 

query = "Do you want to reschedule your service?"
docs=db.similarity_search(query)
docs[0]
```

As a Retriver 

we can also convert the vectorstore into retirver class this allows us to easily use it in other langchain methods with largely work with retriver

```python
retriver = db.as_retriever()
docs=retriver.invoke(query)
docs[0]
```

Similarity Search with score

there are some faisss specfic methods. one of them is similirity_search_with_score, which allows to return not only the documents but also the distance score of the query to them. The Returned distance score is L2 distance therefore, a lower score is better

```python
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
docs_score
```
